```python
import os
import random
import shutil

# Ruta de la carpeta con los archivos originales
carpeta_origen = r'C:\Users\PC\Documents\Dataset_Multiclase\Modelo3\retinopatia'

# Ruta de la carpeta de entrenamiento (train)
carpeta_entrenamiento = r'C:\Users\PC\Documents\Dataset_Multiclase\Modelo3\train\retinopatia'

# Ruta de la carpeta de prueba (test)
carpeta_prueba = r'C:\Users\PC\Documents\Dataset_Multiclase\Modelo3\test\retinopatia'

# Proporción de división (75% - 25%)
proporcion_entrenamiento = 0.75

# Obtener la lista de archivos en la carpeta origen
archivos = os.listdir(carpeta_origen)

# Mezclar aleatoriamente la lista de archivos
random.shuffle(archivos)

# Calcular el índice para dividir la lista en entrenamiento y prueba
indice_division = int(len(archivos) * proporcion_entrenamiento)

# Dividir la lista en archivos de entrenamiento y prueba
archivos_entrenamiento = archivos[:indice_division]
archivos_prueba = archivos[indice_division:]

# Crear la carpeta de entrenamiento (train) si no existe
if not os.path.exists(carpeta_entrenamiento):
    os.makedirs(carpeta_entrenamiento)

# Copiar los archivos de entrenamiento a la carpeta correspondiente
for archivo in archivos_entrenamiento:
    origen = os.path.join(carpeta_origen, archivo)
    destino = os.path.join(carpeta_entrenamiento, archivo)
    shutil.copy(origen, destino)

# Crear la carpeta de prueba (test) si no existe
if not os.path.exists(carpeta_prueba):
    os.makedirs(carpeta_prueba)

# Copiar los archivos de prueba a la carpeta correspondiente
for archivo in archivos_prueba:
    origen = os.path.join(carpeta_origen, archivo)
    destino = os.path.join(carpeta_prueba, archivo)
    shutil.copy(origen, destino)
```

```python
import os
import shutil
import pandas as pd

# Ruta del archivo de Excel
ruta_excel = 'C:/Users/PC/Documents/ODIR/data.xlsx'

# Ruta del directorio donde se encuentran los archivos
directorio_archivos = 'C:/Users/PC/Documents/ODIR-5K/imagenes'

# Leer el archivo de Excel en un DataFrame
dataframe = pd.read_excel(ruta_excel,sheet_name="Normal")

# Iterar sobre cada fila del DataFrame
for index, row in dataframe.iterrows():
    # Obtener el nombre del archivo y la etiqueta
    nombre_archivo = row['imagen']
    etiqueta = row['label']

    # Ruta completa del archivo
    ruta_archivo = os.path.join(directorio_archivos, nombre_archivo)

    # Verificar si el archivo existe
    if os.path.isfile(ruta_archivo):
        # Crear el directorio de la etiqueta si no existe
        directorio_etiqueta = os.path.join(directorio_archivos, etiqueta)
        os.makedirs(directorio_etiqueta, exist_ok=True)

        # Mover el archivo a la carpeta de la etiqueta
        shutil.move(ruta_archivo, directorio_etiqueta)
        print(f"Archivo {nombre_archivo} movido a la carpeta {etiqueta}")
    else:
        print(f"Archivo {nombre_archivo} no encontrado")

print("Proceso completado.")
```

```python
from PIL import Image
import os

# Lista de carpetas que contienen las imágenes en formato JPG
carpetas_jpg = ["normal2", "glau2"]

# Iterar sobre cada carpeta
for carpeta_jpg in carpetas_jpg:
    # Ruta del directorio que contiene las imágenes en formato JPG
    directorio_jpg = os.path.join("C:/Users/PC/Documents/Dataset/", carpeta_jpg)

    # Ruta del directorio donde se guardarán las imágenes en formato PNG
    directorio_png = os.path.join("C:/Users/PC/Documents/Dataset/imagenes", carpeta_jpg)

    # Crear el directorio PNG si no existe
    os.makedirs(directorio_png, exist_ok=True)

    # Obtener la lista de archivos en el directorio JPG
    archivos_jpg = os.listdir(directorio_jpg)

    # Iterar sobre cada archivo JPG
    for archivo_jpg in archivos_jpg:
        # Verificar si el archivo es de formato JPG
        if archivo_jpg.endswith(".jpg"):
            # Ruta completa del archivo JPG
            ruta_jpg = os.path.join(directorio_jpg, archivo_jpg)

            # Abrir la imagen JPG
            imagen = Image.open(ruta_jpg)

            # Generar la ruta y el nombre de archivo para la imagen PNG
            nombre_png = os.path.splitext(archivo_jpg)[0] + ".png"
            ruta_png = os.path.join(directorio_png, nombre_png)

            # Guardar la imagen en formato PNG
            imagen.save(ruta_png, "PNG")
            print(f"Imagen {archivo_jpg} convertida a PNG")

            # Eliminar el archivo JPG
            os.remove(ruta_jpg)
            print(f"Archivo {archivo_jpg} eliminado")

print("Proceso completado.")
```

```python
import os
import random
import shutil

# Ruta del directorio raíz del dataset
directorio_raiz = "C:/Users/PC/Documents/Eye_diseases_dataset/dataset/"

# Ruta del directorio donde se crearán las carpetas "train" y "test"
directorio_destino = "C:/Users/PC/Documents/Eye_diseases_dataset/dataset/"

# Crear las carpetas "train" y "test" si no existen
os.makedirs(os.path.join(directorio_destino, "train"), exist_ok=True)
os.makedirs(os.path.join(directorio_destino, "test"), exist_ok=True)

# Obtener la lista de carpetas originales
carpetas_originales = ["cataratas", "glaucoma", "retinopatia", "normal"]

# Iterar sobre cada carpeta original
for carpeta_original in carpetas_originales:
    # Ruta completa de la carpeta original
    ruta_carpeta_original = os.path.join(directorio_raiz, carpeta_original)

    # Ruta de las subcarpetas "train" y "test" correspondientes
    ruta_carpeta_train = os.path.join(directorio_destino, "train", carpeta_original)
    ruta_carpeta_test = os.path.join(directorio_destino, "test", carpeta_original)

    # Crear las subcarpetas "train" y "test" si no existen
    os.makedirs(ruta_carpeta_train, exist_ok=True)
    os.makedirs(ruta_carpeta_test, exist_ok=True)

    # Obtener la lista de archivos en la carpeta original
    archivos = os.listdir(ruta_carpeta_original)

    # Obtener una lista aleatoria de índices para dividir los archivos en train y test
    indices = list(range(len(archivos)))
    random.shuffle(indices)
    indice_limite = int(0.8 * len(indices))
    indices_train = indices[:indice_limite]
    indices_test = indices[indice_limite:]

    # Mover los archivos a las subcarpetas correspondientes
    for i, indice in enumerate(indices):
        nombre_archivo = archivos[indice]
        ruta_origen = os.path.join(ruta_carpeta_original, nombre_archivo)

        if i in indices_train:
            ruta_destino = os.path.join(ruta_carpeta_train, nombre_archivo)
        else:
            ruta_destino = os.path.join(ruta_carpeta_test, nombre_archivo)

        shutil.move(ruta_origen, ruta_destino)

        print(f"Archivo {nombre_archivo} movido a {ruta_destino}")

print("Proceso completado.")
```

```python
import os
import shutil

# Directorio de origen
directorio_origen = r'C:\Users\PC\Documents\FIVES A Fundus Image Dataset for AI-based Vessel Segmentation'

# Directorio de destino base
directorio_destino_base = r'C:\Users\PC\Documents\FIVES A Fundus Image Dataset for AI-based Vessel Segmentation'

# Obtener la lista de archivos en el directorio de origen
archivos = os.listdir(directorio_origen)

# Mover los archivos a los directorios de destino correspondientes
for archivo in archivos:
    # Obtener el nombre de la carpeta de destino
    nombre_archivo = os.path.splitext(archivo)[0]  # Obtener el nombre del archivo sin extensión
    expresion_despues_del_guion = nombre_archivo.split('_')[1]  # Obtener la expresión después del guión bajo
    
    # Construir la ruta completa del directorio de destino
    directorio_destino = os.path.join(directorio_destino_base, expresion_despues_del_guion)
    
    # Crear el directorio de destino si no existe
    if not os.path.exists(directorio_destino):
        os.makedirs(directorio_destino)
    
    # Construir la ruta completa del archivo de origen y destino
    ruta_origen = os.path.join(directorio_origen, archivo)
    ruta_destino = os.path.join(directorio_destino, archivo)
    
    # Mover el archivo al directorio de destino
    shutil.move(ruta_origen, ruta_destino)
```

```python
import os

# Directorio base que contiene las carpetas
directorio_base = r'C:\Users\PC\Documents\\Dataset_Binario'

# Recorrer las carpetas
for carpeta in os.listdir(directorio_base):
    # Obtener la ruta completa de la carpeta
    ruta_carpeta = os.path.join(directorio_base, carpeta)
    
    # Verificar si es una carpeta
    if os.path.isdir(ruta_carpeta):
        # Obtener la lista de archivos en la carpeta
        archivos = os.listdir(ruta_carpeta)
        
        # Recorrer los archivos y asignar numeración incremental
        for indice, archivo in enumerate(archivos, start=1):
            # Construir la ruta completa del archivo
            ruta_archivo = os.path.join(ruta_carpeta, archivo)
            
            # Obtener el nombre de la carpeta
            nombre_carpeta = carpeta
            
            # Obtener la extensión del archivo
            nombre, extension = os.path.splitext(archivo)
            
            # Construir el nuevo nombre del archivo
            nuevo_nombre = f"{indice}_{nombre_carpeta}{extension}"
            
            # Construir la ruta completa del archivo con el nuevo nombre
            nueva_ruta_archivo = os.path.join(ruta_carpeta, nuevo_nombre)
            
            # Renombrar el archivo
            os.rename(ruta_archivo, nueva_ruta_archivo)
```
